üìå Step 2: Exploratory Data Analysis (EDA)

1Ô∏è‚É£ Univariate Analysis

```python
plt.figure(figsize =(10,5))
sns.histplot(df['TAXI_ID'] , bins=50 , kde=True)
plt.title("Distribution of Taxi Usage")
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x=df['CALL_TYPE'] , palette='viridis')
plt.title('Call Type Distribution')
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x=df['DAY_TYPE'], palette = 'coolwarm')
plt.title('Day Type Distribution')
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='CALL_TYPE' , hue="DAY_TYPE" , data=df)
plt.title('Trip Demand Based on Call Type and Day Type')
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="DAY_TYPE", y="TAXI_ID", data=df)
plt.title("Taxi Distribution Based on Day Type")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="DAY_TYPE", y="hour", data=df)
plt.title("Time of Day for Different Day Types")
plt.show()
```
