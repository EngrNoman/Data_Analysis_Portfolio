```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data set Import
df = pd.read_excel('uncleaned bike sales data.xlsx' )
df.head()
```

```python
df.dropna()
```

```python
df.info()
```

```python
df['Country'].unique()
# df['Country'].value_counts()
```

```python
# Strip Extra Spaces
df['Country'] = df['Country'].str.strip()
```

```python
df['Country'].unique()
```

```python
# Normalize Spaces (Remove Extra Spaces Between Words)
df['Country'] = df['Country'].str.replace(r"\s+" , ' ' , regex = True).str.strip()
```

```python
df['Country'].unique()
```

```python
# Check Duplicated
df[df.duplicated()]
```

```python
#  Remove the Duplicates
df = df.drop_duplicates()
```

```python
df[df.duplicated()]
```

```python
df[df['Unit_Cost']==0]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df = df[df['Unit_Cost'] != 0]
```

```python
df[df['Unit_Cost']==0]
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df['Day'] = df['Day'].astype(int)
```

```python
df["Order_Quantity"] = df["Order_Quantity"].astype(int)
```

```python
df.info()
```

```python
df.to_excel("clean_data.xlsx", index=False)
```
